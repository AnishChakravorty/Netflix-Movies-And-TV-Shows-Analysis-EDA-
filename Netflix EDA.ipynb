## 2. ***Data Wrangling***

<a id='cleaning'></a>
## Data Cleaning

````python
df.head()
````

````python
# Handling missing values

# Replace missing values in 'director' with 'Unknown'
df['director'].fillna('Unknown', inplace=True)

# Replace missing values in 'cast' with 'Unknown'
df['cast'].fillna('Unknown', inplace=True)

# Replace missing values in 'country' with 'Unknown'
df['country'].fillna('Others', inplace=True)

# Replace missing values in 'date_added' with 'Unknown'
df['date_added'].fillna('Unknown', inplace=True)

# Replace missing values in 'rating' with the most frequent value (mode)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
````

````python
df.isnull().sum()
````

````python
df.info()
````

````python
# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
````

````python
# Split 'duration' into 'duration_value' and 'duration_unit'
df[['duration_value', 'duration_unit']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')

# Convert 'duration_value' to numeric
df['duration_value'] = pd.to_numeric(df['duration_value'], errors='coerce')
````

````python
# Convert 'rating' and 'type' to category
df['rating'] = df['rating'].astype('category')
df['type'] = df['type'].astype('category')
````

````python
# Replace "Season" with "Seasons" in the 'duration_unit' column
df['duration_unit'] = df['duration_unit'].replace('Season', 'Seasons')
````

### Data Type Changes

Based on the dataset's structure and the provided data types, the following changes can be made to ensure proper data representation:

1. **`date_added`**: 
    - Current Dtype: `object`
    - Suggested Dtype: `datetime`
    - Reason: Dates should be stored as `datetime` for easier manipulation and analysis.

2. **`release_year`**:
    - Current Dtype: `int64`
    - Suggested Dtype: No change needed.
    - Reason: It is already an integer, which is appropriate for a year.

3. **`duration`**:
    - Current Dtype: `object`
    - Suggested Dtype: Split into two columns:
      - `duration_value` (numeric): Extract the numeric part of the duration.
      - `duration_unit` (categorical): Extract the unit (e.g., "min", "Seasons").
    - Reason: Storing duration as separate numeric and categorical columns allows for better analysis.

4. **`rating`**:
    - Current Dtype: `object`
    - Suggested Dtype: `category`
    - Reason: Ratings have a limited set of unique values, so using `category` reduces memory usage and improves performance.

5. **`type`**:
    - Current Dtype: `object`
    - Suggested Dtype: `category`
    - Reason: The column contains only two unique values ("Movie" and "TV Show"), so `category` is more efficient.

6. **`country`**, **`director`**, **`cast`**, **`listed_in`**:
    - Current Dtype: `object`
    - Suggested Dtype: No change needed.
    - Reason: These columns contain textual data, which is appropriate as `object`.

### Summary of Changes:
- Convert `date_added` to `datetime`.
- Split `duration` into `duration_value` (numeric) and `duration_unit` (categorical).
